```python
import numpy as np
import pandas as pd
movies=pd.read_csv(r"C:\Users\admin\OneDrive\Desktop\New folder\tmdb_5000_movies.csv")
credits=pd.read_csv(r"C:\Users\admin\OneDrive\Desktop\New folder\tmdb_5000_credits.csv")
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies['original_language'].value_counts()
```

```python
movies = movies[['id','title','genres','keywords','cast','crew','release_date','runtime','overview']]
```

```python
movies
```

```python
movies.isnull().sum()
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3 :
            L.append(i['name']) 
            counter+=1
        else :
            break
        
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+ movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
movies
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit(new_df['tags'])
```

```python
vectors=cv.transform(new_df['tags'])
```

```python
vectors.toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
